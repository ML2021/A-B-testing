```python
mu_a=a.mean() #mean
v_a=np.var(a) #variance
sigma = math.sqrt(v_a) 
v_a,mu_a,sigma, np.std(a)
```

```python
mu_b=b.mean() #mean
v_b=np.var(b) #variance
sigma_b = math.sqrt(v_b) 
v_b,mu_b,sigma_b, np.std(b)
```

```python
#Lets calculate the t test
```

```python
var_a = a.var(ddof=1)
var_a 
```

```python
var_b = b.var(ddof=1)
var_b 
```

```python
s= np.sqrt((var_a + var_b)/2)
print("standard deviation s=  sqrt variance a+ variance b  / N ")
s
```

```python
# t test  t value
t = (a.mean() -b.mean()) / (s *np.sqrt (2.0/N))
t
```

```python
#degree of freedom
df=2*N-2
df
```

```python
#calculate p value .. if its lesss that .005 there IS A DIFFERENCE
p = 1- stats.t.cdf(t, df=df)
p = 2*p  #Must*2 as its 2 tail
p
```
